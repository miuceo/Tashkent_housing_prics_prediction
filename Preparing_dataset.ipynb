```python
df['size'] = df['size'].replace("Площадьземли:1сот", "100")
df["price"] = df["price"].replace("Договорная", 0)
```

```python
df[['price' , 'size']] = df[['price', 'size']].astype(float)
```

```python
df['district'] = df['district'].replace({'Юнусабадский': 'Yunusobod',
    'Яккасарайский': 'Yakkasaroy',
    'Чиланзарский': 'Chilonzor',
    'Учтепинский': 'Uchtepa',
    'Мирзо-Улугбекский': 'Mirzo-Ulugʻbek',
    'Яшнободский': 'Yashnobod',
    'Мирабадский': 'Mirabad',
    'Шайхантахурский': 'Shayxontohur',
    'Олмазорский': 'Olmazor',
    'Сергелийский': 'Sergeli',
    'Янгихаётский': 'Yangihayot',
    'Бектемирский': 'Bektemir'})
```

```python
df["price"] = df["price"].replace(0, np.nan)
df['psp'] = df['price'] / df['size']
df_gr = df.groupby("district")['psp'].mean()
```

```python
df_gr
```

```python
for i in range(len(df)):
    if pd.isna(df['price'][i]):
        for j in df_gr.index:
            if df['district'][i] == j:
                df['price'][i] = df_gr[j] * df['size'][i]
                df['psp'][i] = df_gr[j]
```
